```python
pd.options.display.max_columns=100

data = pd.read_json('../data/twint_tweets.json', lines=True)
print('SHAPE',data.shape)
data.head()
```

```python
twitter = data[['id','tweet','language', 'date', 'time','user_id','name', 'username', 'retweets_count', 'replies_count', 'likes_count']]
twitter=twitter[twitter['language']=='es']
twitter['date']=twitter['date'].astype(str)

twitter.to_csv('data/twitter.csv', index=False)
twitter.head(1)
```

```python
twitter_tweets = twitter[['id','tweet','language', 'date', 'time', 'retweets_count', 'replies_count', 'likes_count','user_id']]

# twitter_tweets.to_csv('data/twitter_tweets.csv', index=False)

twitter_tweets.head(1)
```

```python
twitter_users = data[['user_id','name', 'username']]

# twitter_users.to_csv('data/twitter_users.csv', index=False)

twitter_users.head(1)
```

```python
twitter_users.head(1)
```

```python
# Creamos la base de datos y sus tablas
import sqlite3
sql = sqlite3.connect('data/DB_twitter.db')
cursor = sql.cursor()

lista_valores_tweets = twitter_tweets.values.tolist()
lista_valores_users = twitter_users.values.tolist()

cursor.execute('''CREATE TABLE users(user_id INT PRIMARY KEY, name TEXT, username TEXT)''')
cursor.execute('''CREATE TABLE tweets(id TEXT, tweet TEXT, language TEXT, date DATE, time TEXT, retweets_count INT, replies_count INT, likes_count INT, user_id INT, FOREIGN KEY(user_id) REFERENCES users(user_id))''')

cursor.executemany("INSERT or IGNORE INTO users VALUES (?,?,?)", lista_valores_users)
cursor.executemany("INSERT or IGNORE INTO tweets VALUES (?,?,?,?,?,?,?,?,?)", lista_valores_tweets)

sql.commit()
sql.close()
```
